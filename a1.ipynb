# Problem 3

For this problem, you need to write a function **plotData3** to make a figure with two subplots.  One is a bar plot, and the other is a histogram plot. The bar plot can conveniently support users to visualize how many accounts are paid and how many are unpaid. The specific numbers of accounts for each type should be presented. The histogram plot allows us discover the underlying frequency distribution of salary data associated with the two types of data scientists (paid and unpaid).  Your plot should be similar to the figure that can be displayed below by running the code cell.

```python
def plotData3(data):
    """
    data: a list of data scientist tuples
    The function plots the the data scientists as required in the problem
    """
    # Separate data into paid and unpaid
    paid = [(tenure, salary) for (tenure, salary, account) in data if account == 1]
    unpaid = [(tenure, salary) for (tenure, salary, account) in data if account == 0]

    # Count how many paid vs. unpaid
    num_paid = len(paid)
    num_unpaid = len(unpaid)

    # Extract salaries for paid and unpaid 
    paid_salaries   = [s for (_, s) in paid]
    unpaid_salaries = [s for (_, s) in unpaid]

    # Create a 1 row, 2 column figure
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    
    # Positions for the bars
    x_positions = [0, 1]
    # Heights
    heights = [num_paid, num_unpaid]
    # Labels
    labels = ['Paid', 'Unpaid']

    ax1.bar(x_positions, heights, color=['green', 'red'], alpha=0.7)
    ax1.set_xticks(x_positions)
    ax1.set_xticklabels(labels)
    ax1.set_ylabel('Number of Accounts')
    ax1.set_title('Paid vs. Unpaid Accounts')

    # Annotate bars with their counts
    for i, v in enumerate(heights):
        ax1.text(i, v + 0.5, str(v), ha='center', fontweight='bold')

    # Salaries Histogram 
    
    # Plot side-by-side or overlapping histograms
    ax2.hist([paid_salaries, unpaid_salaries],
             bins=10,  # You can adjust the bin count as needed
             color=['green', 'red'],
             alpha=0.7,
             label=['Paid', 'Unpaid'])
    ax2.set_xlabel('Salary')
    ax2.set_ylabel('Frequency')
    ax2.set_title('Salary Distribution')
    ax2.legend()

    # Adjust layout and show the combined figure
    plt.tight_layout()
    plt.show()


    pass

plotData3(data)
```

For this problem, you need to write a function **barCommunityData** make a bar plot that shows numbers of counties in states.  Your plot should be similar to the figure that can be displayed below by running the code cell that has the Image object.  To approach this problem, you need to write the function definition that includes statements to parse the data from the provided file named "county.csv". For this assignment, you are **not** allowed to use pandas library to process the data file.   You are required to read the file row by row and extract the data to address the question.  In your plot, the specific numbers of counties in the states should be presented.

For the reference result plot, it shows the numbers of counties in Arizona, California, and Texas. However, your **barCommunityData** applications should allow any other states on the second argument.  

```python
def barCommunityData(states):
    """
    states: a list of strings representing the states to count the counties
    """
    # Count how many countes in each state
    state_counties = {state: 0 for state in states}

    # open the csv file
    with open('county.csv', 'r') as infile:
        # Read the header
        header = infile.readline()
        # Read the rest of the file
        for line in infile:
            # Split the line
            parts = line.split(',')
            # Get the state
            state = parts[1]
            # Check if the state is in the list
            if state in states:
                # Increment the count
                state_counties[state] += 1
    # Prepare the data for plotting
    states = list(state_counties.keys())
    counts = list(state_counties.values())

    # Create a bar plot
    plt.figure(figsize=(10, 5))
    colors = ['red', 'green', 'blue']
    # Label x axis with states names
    x_expostions = range(len(states))
    plt.bar(states, counts, color=colors, alpha=0.7)
    plt.xticks(x_expostions, states)
    plt.ylabel('')
    plt.title('Number of Counties in States')
    # remove grid lines 
    plt.grid(False)

    # Build the info string for the key
    info_str = "\n".join([f"{st}: {ct}" for st, ct in zip(states, counts)])
   
# Add the text box in the top-left, relative to axes coordinates
    plt.text(
        0.02, 0.95,            # X and Y position (0.0 = left/bottom, 1.0 = right/top of axes)
        info_str, 
        transform=plt.gca().transAxes,  # place relative to the axes, not data
        fontsize=12, 
        ha='left', 
        va='top',
        bbox=dict(facecolor='white', alpha=0.5, boxstyle='round,pad=0.5')
    )
        
    # Show the plot
    plt.show()
    pass

barCommunityData(['Arizona', 'California', 'Texas']) #test code
```
